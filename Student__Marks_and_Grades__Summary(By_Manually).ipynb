```python
import matplotlib.pyplot as plt

def cal_grade(average):
    if average >= 85:
        return 'A'
    elif average >= 70:
        return 'B'
    elif average >= 55:
        return 'C'
    elif average >= 40:
        return 'D'
    else:
        return 'F'

# Input: number of students
n = int(input("Enter number of students: "))
students = []

# Collect student data
for i in range(n):
    print(f"\nEnter details for student {i+1}:")
    name = input("Name: ")
    try:
        m1 = float(input("Marks in Subject 1: "))
        m2 = float(input("Marks in Subject 2: "))
        m3 = float(input("Marks in Subject 3: "))
    except ValueError:
        print("Invalid input! Marks must be numbers.")
        exit()

    total = m1 + m2 + m3
    average = total / 3
    grade = cal_grade(average)

    students.append({
        "name": name,
        "total": total,
        "average": average,
        "grade": grade
    })

# Display Report
print("\nStudent Summary Report:")
print("-" * 50)
print(f"{'Name':<12}{'Total':<10}{'Average':<10}{'Grade'}")
print("-" * 50)

total_class_avg = 0
topper = students[0]

for s in students:
    print(f"{s['name']:<12}{s['total']:<10.2f}{s['average']:<10.2f}{s['grade']}")
    total_class_avg += s['average']
    if s['average'] > topper['average']:
        topper = s

class_avg = total_class_avg / n
print(f"\nClass Average: {class_avg:.2f}")
print(f"Topper: {topper['name']} with Average {topper['average']:.2f} and Grade {topper['grade']}")

# Visualization with matplotlib
names = [s['name'] for s in students]
averages = [s['average'] for s in students]

plt.figure(figsize=(8, 5))
bars = plt.bar(names, averages, color='skyblue')
plt.axhline(class_avg, color='red', linestyle='--', label='Class Average')

# Highlight topper in green
topper_index = names.index(topper['name'])
bars[topper_index].set_color('limegreen')

plt.xlabel("Student Names")
plt.ylabel("Average Marks")
plt.title("Average Marks per Student")
plt.legend()
plt.tight_layout()
plt.show()
```
